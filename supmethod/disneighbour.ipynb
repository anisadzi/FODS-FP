```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import classification_report
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('disCleaned.csv')
df.info()
```

```python
#taking year 2000 mean

df['Index'] = 'Higher'
df.loc[df['2000']<= 2577	,'Index'] ="Lower" 
# X = df.loc[df['2000']<= 2577	,'Index'] ="Minor" #features
X = df.iloc[0:34]#features
y = df['Index']#target
print(y)
```

```python
X=pd.get_dummies(X)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=40)
```

```python
knn = KNeighborsClassifier()
kn_class = KNeighborsClassifier()
kn_class.fit(X_train, y_train)
predict = kn_class.predict(X_test)
predict
```

```python
def Report(y_test, y_pred):
    print("Report : ",
    classification_report(y_test, y_pred))
```

```python
print("Trained acc: ", accuracy_score(y_train, kn_class.predict(X_train)))
print("Test accuracy:", accuracy_score(y_test, predict))
print("Confusion matrix train: ", "\n", confusion_matrix(y_train, kn_class.predict(X_train)))
print("Confusion matrix test: ", "\n",confusion_matrix(y_test, predict))
```

```python
print(Report(y_train, kn_class.predict(X_train)))
print("---------------------------------------------------------")
print(Report(y_test, predict))
```
